```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
model_data = pd.read_csv('./dataset00.csv', index_col=0)
```

```python
model_data.head()
```

```python
model_data['SalePrice'].describe()
```

```python
print("Skewness: %3.2f , Kurtosis: %3.2f" % (model_data['SalePrice'].kurt() , model_data['SalePrice'].skew()))
```

```python
corrmat = model_data.corr()
print (corrmat)
```
